# Analysis of experiment #4: Randomness of the pricing heuristic

## Read data

```python
# Read csv

import pandas as pd

df = pd.read_csv("stats4.csv")

df
```

```python
df.columns
```

## Summary

```python
df3 = df.groupby(["solver"]).agg(
    {
        "state": [("feasible", lambda x: (x == "FEASIBLE").sum()), ("optimal", lambda x: (x == "OPTIMAL").sum())],
        "time": [("avg", "mean")],
        "rootNCallsHeur": [("avg", "mean")],
        "rootNColsHeur": [("avg", "mean")],
        "rootNCallsExact": [("avg", "mean")],
    }
).reset_index()
df3
```
