# Analysis of experiment #2
## Number of repetitions of semigreedy heuristic

## Read data

```python
# Read csv

import pandas as pd

df = pd.read_csv("stats2.csv")

# Some renames to improve readability
df["solver"] = df["solver"].str.replace("v2", "deg")
df["solver"] = df["solver"].str.replace("v3", "edg")

# For the unsolved instances,
# --> write NaN in ub, bestTime, bestIter
df.loc[df.state != "FEASIBLE", "value"] = float("nan")
df.loc[df.state != "FEASIBLE", "bestTime"] = float("nan")
df.loc[df.state != "FEASIBLE", "bestIter"] = float("nan")

# Rename some columns
df = df.rename(columns={'state': 'solved'})

# Separate a column for the variant and the repetitions
df["variant"] = df.solver.apply(lambda s: s.split("_")[2])
df["repetitions"] = df.solver.apply(lambda s: int(s.split("_")[3][1:]))

df
```

## Summary

```python
df2 = df.groupby(["solver", "variant", "repetitions"]).agg(
    {
        "solved": [("count", lambda x: (x == "FEASIBLE").sum())],
        "value": [("sum", "sum")],
        "totalTime": [("avg", "mean")],
        "bestIter": [("avg","mean"),("max", "max")]
    }
).reset_index()
df2 = df2.sort_values(["variant","repetitions"])
df2
```
